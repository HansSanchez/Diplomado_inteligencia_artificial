## Parte 3: Consolidación de información
#### Web Scraping a "Infobae" para investigación de informes referentes a riegos pre electorales

```python
# Importación de librerias
import pandas as pd

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
            ## encoding='utf-8' es por los caracteres especiales que puede tener los datos
        self.df_1 = pd.read_csv(r'violencia_política_infobae_parte_2.csv', encoding='utf-8') 
        self.df_2 = pd.read_csv(r'manipulación_de_información_infobae_parte_2.csv', encoding='utf-8')
        self.df_3 = pd.read_csv(r'fraude_electoral_infobae_parte_2.csv', encoding='utf-8')
        self.df_4 = pd.read_csv(r'financiamiento_ilegal_de_campañas_infobae_parte_2.csv', encoding='utf-8')
        self.df_5 = pd.read_csv(r'compra_de_votos_infobae_parte_2.csv', encoding='utf-8')
```

```python
# Instancia del objeto
IC = ImportClass()
```

```python
# Agregar la nueva columna con un valor predeterminado
IC.df_1['types'] = "Violencia política"
```

```python
# Agregar la nueva columna con un valor predeterminado
IC.df_2['types'] = "Manipulación de información"
```

```python
# Agregar la nueva columna con un valor predeterminado
IC.df_3['types'] = "Fraude electoral"
```

```python
# Agregar la nueva columna con un valor predeterminado
IC.df_4['types'] = "Financiamiento ilegal"
```

```python
# Agregar la nueva columna con un valor predeterminado
IC.df_5['types'] = "Compra de votos"
```

```python
# Consolidación en un único dataframe de todas las noticias para su procesamiento
df_consolidated = pd.concat([IC.df_1, IC.df_2, IC.df_3, IC.df_4, IC.df_5])
```

```python
# Limpieza del dataframe de noticias iguales
df_consolidated.drop_duplicates(subset=['paragraphs'])
```

```python
# Construcción del archivo final consolidado con las noticias de todos los temas
df_consolidated.to_csv("noticias_sobre_riesgos_pre_electorales_consolidadas_de_infobae.csv")
```
