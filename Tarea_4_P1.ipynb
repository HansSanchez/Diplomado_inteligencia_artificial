2. Escriba una clase iterable que tenga la siguiente funcionalidad
- Al instanciar un objeto, lee el archivo de datos y lo coloca en un dataframe interno en la clase.
- Cada vez que se le pidan datos, debe regresar los dos siguientes datos.
- Entregue un error si los datos se han agotado


```python
# Importación de libreria
import pandas as pd

# Definición de la clase
class IterableClass:
    
    # Función inicial o constructor
    def __init__(self, list_values = None): # None es para que sea opcional
        self.list_values = list_values
        self.dateframe = pd.read_csv(r'Recursos/Iris.csv') # lectura del archivo
        self.i = 0
        
    # Retornar un iterador
    def __iter__(self):
        return self
    
    # Retorna un elemento del contenedor
    def __next__(self):
        if self.i >= len(self.list_values):
            print("\nNo hay más valores a recorrer... ⬇")
            raise StopIteration
            
        item = self.list_values[self.i]
        self.i += 1
        return item    
```

```python
# Instancia del objeto
IC = IterableClass()

# Acceso a la lectura del archivo Iris.csv
IC.dateframe
```

```python
# Definición de variable para mejor visualización
columns = IC.dateframe.columns

# Ciclo para impresión de valores
for column in columns:
    print(column, end=", ")
```

```python
# Captura de la columna de interés
column = input("Qué columna quieres ver: ") 

# Filtro en el dataframe según columna
values = IC.dateframe[column]

# Definición de array
list_values = []

# Ciclo para agregar todos los valores en un array para posterior iteración
for value in values: 
    list_values.append(value)
print('\nHecho!\n')

# Comprobación de valores añadidos al array
list_values
```

```python
# Definición del objeto de la clase iterable
list_iterator = IterableClass(list_values)

# Ciclo para recorrer los valores y forzar el error
for i in range(len(list_values) + 1):
    print(next(list_iterator))
```
