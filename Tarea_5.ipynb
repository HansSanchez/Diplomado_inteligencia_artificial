1. Cree dos archivos de tipo csv. En el primero debe colocar la información de 100 personas (clientes): id, cedula, nombre, apellido, número de cuenta.  En el segundo debe colocar transacciones de cada cliente. Por ejemplo (id= 3456, fecha = 29/01/2023, valor = -50_000). En total  mil transacciones.

2. Construya una función que lea el primer archivo y los suba a Pandas.
3. Construya una función  que haga lo mismo con el segundo archivo.
4.  Construya una función que reciba los dos objetos de Pandas y cree un diccionario, en donde hay solamente un registro por cada cliente. Por ejemplo ( esto es solamente un ejemplo)  {id= 2345, nombre='Luis Vazquez', cedula=20456912, transacciones=[ {fecha: 0/12/2021, valor: 45_000}, {fecha: 15/02/2023; valor: -70_000}, ...]}
5. Convierta el diccionario en un objeto json.
6. Guarde el objeto json como un archivo json.
7. Con base en todo lo anterior construya una clase, siguiendo las mejores prácticas que se verán en clase, que implemente toda la solución.

```python
# Importación de librerias
import pandas as pd
import json as js

# Definición de la clase
class ImportClass:
    
    # Función inicial o constructor
    def __init__(self):
        # lectura de los archivos
            ## encoding='ISO-8859-1' es por los caracteres especiales que puede tener los datos
            ## sep=';' es para especificarle con que caracter queremos que nos reconozca la separación
        self.dateframe_clientes = pd.read_csv(r'Recursos/tarea_5_clientes.csv', encoding='ISO-8859-1', sep=';') 
        self.dateframe_movimientos = pd.read_csv(r'Recursos/tarea_5_movimientos.csv', encoding='ISO-8859-1', sep=';')
    
    # Función para crear un diccionario
    def createDictionary(self, dateframe_clientes = None, dateframe_movimientos = None):
        try:
             # Unir los dos objetos de Pandas en uno solo
                ## drop_duplicates(subset=['cliente_id']) es para eliminar los diferentes movimientos de un cliente en base a su id
            datos = pd.merge(dateframe_clientes, dateframe_movimientos.drop_duplicates(subset=['cliente_id']), on='cliente_id')

            # Crear un diccionario para almacenar la información de cada cliente
            diccionario = {}

            # Iterar sobre los registros en el objeto unido
            for _, registro in datos.iterrows():
                
                # Obtener el id del cliente y crear un registro si aún no existe
                cliente_id = registro['cliente_id']
                
                #Validación del id en el diccionario
                if cliente_id not in diccionario:
                    diccionario[cliente_id] = {
                        'cliente_id': cliente_id,
                        'cedula': registro['cedula'],
                        'nombre': registro['nombre'],
                        'apellido': registro['apellido'],
                        'numero_de_cuenta': registro['numero_de_cuenta'],
                        'transacciones': []
                    }

                # Agregar la información de la transacción al registro del cliente
                diccionario[cliente_id]['transacciones'].append({
                    'fecha': registro['fecha'],
                    'valor': registro['valor'],
                    'cliente_id': cliente_id
                })
            
            # Retorno de los valores en lista del diccionario
            return list(diccionario.values())
        except Exception as e: 
            print("Revisa el error => ", e) # Impresión del error personalizado
```

```python
# Instancia del objeto
IC = ImportClass()

# Acceso a la lectura del archivo Iris.csv
IC.dateframe_clientes
```

```python
# Acceso a la lectura del archivo Iris.csv
IC.dateframe_movimientos
```

```python
# Variable para definición del Dataframe - Clientes
dfc = IC.dateframe_clientes

# Variable para definición del Dataframe - Movimientos
dfm = IC.dateframe_movimientos

# Llamado de la función para la creación de un diccionario
IC.createDictionary(dfc, dfm)
```

```python
# Asignación de la función a una variable para uso posterior
datos = IC.createDictionary(dfc, dfm)
```

```python
# Convertimos el diccionario en un objeto JSON
json_datos = js.dumps(datos)

# Imprimimos el objeto JSON
json_datos
```

```python
# Guardamos el objeto JSON en un archivo
    ## la "w" es el parametro para written (escritura)
with open(r'Recursos/datos.json', 'w') as archivo:
    js.dump(json_datos, archivo)
```
